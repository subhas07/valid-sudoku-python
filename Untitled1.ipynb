```python
#VALID SUDOKU
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
#FIRST WE WILL CHECK THE SUM OF THE ROWS AS IT IS 45 OR NOT.
r=np.sum(s,axis=1)
for i in r:
    if i!=45:
        print("This sudoku is invalid")
        break
else:
    print("this is a valid sudoku")
```

```python
#NOW WE WILL CHECK THE SUM OF THE COLUMNS AS IT IS 45 OR NOT.
c=np.sum(s,axis=0)

for i in c:
    if i!=45:
        print("this sudoku is invalid")
        break
else:
    print("this is a valid sudoku")
```

```python
#NOW WE WILL CHECK WHETHER THE SUM OF 3x3 GRIDS ARE 45 OR NOT
for i in range(0,9,3):
    for j in range(0,9,3):
        n=s[i:i+3,j:j+3]
        print(n.sum())
```
