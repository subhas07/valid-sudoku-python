```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
r=np.sum(s,axis=1)
if np.all(r==45):
    print("this sudoku is correct for rows")
else:
    print("a invalid sudoku")
```

```python
c=np.sum(s,axis=0)
for i in c:
    if i!=45:
        print("this is a invalid sudoku")
        break
else:
    print("this sudoku is valid for columns also")
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        n=s[i:i+3,j:j+3]
        print(n.sum())
```

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
np.mean(data[:,0])
```

```python
#Find the average math marks of students.
np.mean(data[:,1])
```

```python
#Extract Math marks of all students.
math_marks=data[:,1]
print(math_marks)
```

```python
#Find the highest Science mark.
science_marks=data[:,2]
highest=np.max(science_marks)
print(highest)
```

```python
#Get details of the student who scored more than 90 in Math.
import numpy as np

data = np.array([
    [18, 85, 78],
    [19, 92, 88],
    [17, 76, 95],
    [18, 65, 70],
    [20, 90, 85]
])

data[data[:,1] > 90]
```

```python
#Get details of the student who scored more than 90 in science.
data[data[:,2]>90]
```

```python
#Increase Math marks of all students by 5.
data_1=data[:,1]+5
print(data_1)
```

```python
#Increase age of all students by 15.
age_in=data[:,0]+15
print(age_in)
```

```python
#Find how many students are younger than 19.
#np.sum(data[:,0]<19)
int(np.sum(data[:,0] < 19))
```

```python
#Calculate the average marks in each subject (column-wise mean).
avg=np.mean(data[:,1:],axis=0)
print(avg)
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```

```python
#Replace all Science marks < 75 with 0.
data[data[:, 2] < 75, 2] = 0
print(data)
```
